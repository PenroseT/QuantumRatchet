```julia
using LinearAlgebra
using CairoMakie
using LaTeXStrings
```

```julia
function rhosteady(mx::Float64, my::Float64, mz::Float64,
                   hx::Float64, hz::Float64, γ1::Float64,
                   γ2::Float64, τ::Float64)
    
    Γ1 = γ1*τ
    Γ2 = γ2*τ
    ηx = hx*τ
    ηz = hz*τ
    η = sqrt(ηx^2+ηz^2)
    hnorm = sqrt(hx^2+hz^2)
    mdoth=(mx*hx+mz*hz)/hnorm
    
    det = 2*η^2*(1-mdoth^2)/Γ1+4*ηz^2+4*ηx^2*Γ2/Γ1+(1-Γ2/Γ1)*mz^2+Γ2/Γ1+2*Γ2*(1+mz^2)+4*Γ2^2+2*Γ2^2*(1-mz^2)/Γ1
    
    rhoxnum = -mx*mz-2*η*my*mdoth+2*Γ2*mx*mz-4*ηx*ηz
    rhoynum = -my*mz-2*η*mx*mdoth-2*Γ2*my*mz+2*ηx*(1-2*Γ2)
    rhoznum = -mz^2-2*Γ2*(1+mz^2)-4*(ηz^2+Γ2^2)
    
    rho = [rhoxnum/det, rhoynum/det, rhoznum/det]
    
    return rho
end
```

```julia
θ = -acos(1/sqrt(3))
ϕ = π/4
mx = sin(θ)*cos(ϕ)
my = sin(θ)*sin(ϕ)
mz = cos(θ)
γ2 = 0.1
γ1 = 0.01
hx = 0.1
hz = 1.0
hnorm = sqrt(hx^2+hz^2)
τm = 3.0
Δs = [-10.0:0.01:10.0; ]
rhoys = [rhosteady(mx, my, mz, hx, Δ/2, γ1, γ2, τm)[1] for Δ in Δs]


lines(Δs, rhoys, linewidth=2.0)
```

```julia
θ = acos(1/sqrt(3))
ϕ = π/4
mx = sin(θ)*cos(ϕ)
my = sin(θ)*sin(ϕ)
mz = cos(θ)
γ2 = 0.1
γ1 = 0.01

hxs = [1e-3, 1e-1, 0.5]
Δs = [-3.01:0.01:3.01; ]
τms = [1e-7:1e-6:1e-5; 1e-5:1e-5:1e-3; 1e-3:1e-3:1.0; 1.0:0.2:100.0; 100.0:10.0:1000.0; 1e3:1e3:1e5; 1e5:1e5:1e7; ]
#τms = [100.0:10.0:10000.0; ]
rhoxs = Array{Float64}(undef, length(Δs), length(τms), length(hxs))
rhoys = Array{Float64}(undef, length(Δs), length(τms), length(hxs))
rhots = Array{Float64}(undef, length(Δs), length(τms), length(hxs))
rhozs = Array{Float64}(undef, length(Δs), length(τms), length(hxs))


for k in eachindex(hxs)
    for i in eachindex(τms)
        for j in eachindex(Δs)
            rhoxs[j, i, k] = rhosteady(mx, my, mz, hxs[k], Δs[j]/2, γ1, γ2, τms[i])[1]
            rhoys[j, i, k] = rhosteady(mx, my, mz, hxs[k], Δs[j]/2, γ1, γ2, τms[i])[2]
            rhozs[j, i, k] = rhosteady(mx, my, mz, hxs[k], Δs[j]/2, γ1, γ2, τms[i])[3]
            rhots[j, i, k] = sqrt(rhoxs[j, i, k]^2+rhoys[j, i, k]^2)
        end
    end
end
```

```julia
f = Figure(size=(2200, 2000), fonts = (; regular = "cm", weird = "Blackchancery"))

columnlabels = [L"Ω=10^{-3}", L"Ω=0.1", L"Ω=0.5"]
rowlabels = [L"\langle M_{\mathrm{x}}\rangle", L"\langle M_{\mathrm{y}}\rangle", L"\langle |M_{\perp}|\rangle", L"\langle M_{\mathrm{z}}\rangle"]

for k in eachindex(hxs)
    ax = Axis(f[1, 2*k-1], yscale=log10, xticklabelsize=40, yticklabelsize=40, titlesize=60, ylabelsize=60, title=columnlabels[k], ylabel="$(k==1 ? rowlabels[1] : "")")
    surf = contourf!(Δs, τms, rhoxs[:, :, k], colormap=:coolwarm,
           levels=range(-maximum([-minimum(rhoxs[:, :, k]), maximum(rhoxs[:, :, k])]), maximum([-minimum(rhoxs[:, :, k]), maximum(rhoxs[:, :, k])]), length=15)) 
    Colorbar(f[1, 2*k], surf, ticklabelsize=40)
end

for k in eachindex(hxs)
    ax = Axis(f[2, 2*k-1], yscale=log10, xticklabelsize=40, yticklabelsize=40, ylabelsize=60, ylabel="$(k==1 ? rowlabels[2] : "")")
    surf = contourf!(Δs, τms, rhoys[:, :, k], colormap=:coolwarm,
        levels=range(-maximum([-minimum(rhoys[:, :, k]), maximum(rhoys[:, :, k])]), maximum([-minimum(rhoys[:, :, k]), maximum(rhoys[:, :, k])]), length=15)) 
    Colorbar(f[2, 2*k], surf, ticklabelsize=40)
end

for k in eachindex(hxs)
    ax = Axis(f[3, 2*k-1], yscale=log10, xticklabelsize=40, yticklabelsize=40, ylabelsize=60, ylabel="$(k==1 ? rowlabels[3] : "")")
    surf = contourf!(Δs, τms, rhots[:, :, k], colormap="Greys",
        levels=range(0.0, maximum([-minimum(rhots[:, :, k]), maximum(rhots[:, :, k])]), length=15))
    Colorbar(f[3, 2*k], surf, ticklabelsize=40)
end

for k in eachindex(hxs)
    ax = Axis(f[4, 2*k-1], yscale=log10, xticklabelsize=40, yticklabelsize=40, ylabelsize=60, xlabelsize=60,
        ylabel="$(k==1 ? rowlabels[4] : " ")", xlabel="$(k==2 ? "Δ" : " ")")
    surf = contourf!(Δs, τms, rhozs[:, :, k], colormap=:coolwarm,
        levels=range(-maximum([-minimum(rhozs[:, :, k]), maximum(rhozs[:, :, k])]), maximum([-minimum(rhozs[:, :, k]), maximum(rhozs[:, :, k])]), length=15)) 
    Colorbar(f[4, 2*k], surf, ticklabelsize=40)
end

f
```

```julia
θ = acos(1/sqrt(3))
ϕ = π/4
mx = sin(θ)*cos(ϕ)
my = sin(θ)*sin(ϕ)
mz = cos(θ)
γ2 = 0.1
γ1 = 0.01

τs = [1e-2, 0.5, 5.0, 500.0]
Δs = [-7.51:0.01:7.51; ]
#τms = [1e-7:1e-6:1e-5; 1e-5:1e-5:1e-3; 1e-3:1e-3:1.0; 1.0:0.2:100.0; 100.0:10.0:1000.0; 1e3:1e3:1e5; 1e5:1e5:1e7; ]
#τms = [100.0:10.0:10000.0; ]
hx = 0.1

rhoxs = Array{Float64}(undef, length(Δs), length(τs))
rhoys = Array{Float64}(undef, length(Δs), length(τs))
rhots = Array{Float64}(undef, length(Δs), length(τs))
rhozs = Array{Float64}(undef, length(Δs), length(τs))


for j in eachindex(τs)
    for i in eachindex(Δs)
        rhoxs[i, j] = rhosteady(mx, my, mz, hx, Δs[i]/2, γ1, γ2, τs[j])[1]
        rhoys[i, j] = rhosteady(mx, my, mz, hx, Δs[i]/2, γ1, γ2, τs[j])[2]
        rhozs[i, j] = rhosteady(mx, my, mz, hx, Δs[i]/2, γ1, γ2, τs[j])[3]
        rhots[i, j] = sqrt(rhoxs[i, j]^2+rhoys[i, j]^2)
    end
end
```

```julia
f = Figure(size=(2500, 2000), fonts = (; regular = "cm", weird = "Blackchancery"))

columnlabels = [L"\tau=10^{-2}", L"\tau=0.5", L"\tau=5.0", L"\tau=500.0"]
rowlabels = [L"\langle M_{\mathrm{x}}\rangle", L"\langle M_{\mathrm{y}}\rangle", L"\langle |M_{\perp}|\rangle", L"\langle M_{\mathrm{z}}\rangle"]

for j in eachindex(τs)
    ax = Axis(f[1, j], xticklabelsize=40, yticklabelsize=40, titlesize=60, ylabelsize=60, title=columnlabels[j], ylabel="$(j==1 ? rowlabels[1] : "")")
    line = lines!(Δs, rhoxs[:, j], linewidth=4.0)
end

for j in eachindex(τs)
    ax = Axis(f[2, j], xticklabelsize=40, yticklabelsize=40, titlesize=60, ylabelsize=60, ylabel="$(j==1 ? rowlabels[2] : "")")
    line = lines!(Δs, rhoys[:, j], linewidth=4.0)
end

for j in eachindex(τs)
    ax = Axis(f[3, j], xticklabelsize=40, yticklabelsize=40, titlesize=60, ylabelsize=60, ylabel="$(j==1 ? rowlabels[3] : "")")
    line = lines!(Δs, rhots[:, j], linewidth=4.0)
end

for j in eachindex(τs)
    ax = Axis(f[4, j], xticklabelsize=40, yticklabelsize=40, titlesize=60, ylabelsize=60, ylabel="$(j==1 ? rowlabels[4] : "")")
    line = lines!(Δs, rhozs[:, j], linewidth=4.0)
end

f
```

```julia
module sm
    include("SpinMeasurement.jl")
end

using Plots
using LaTeXStrings
using LinearAlgebra
using CairoMakie
```

```julia
hx=0.5
hy=0.1
hz=1.0
hnorm=sqrt(hx^2+hy^2+hz^2)

θ=acos(1/sqrt(3))
ϕ=π/4

mx=sin(θ)*cos(ϕ)
my=sin(θ)*sin(ϕ) 
mz=cos(θ) 

γ0 = 1e-2

γ0s=[1e-3, 1e-2, 1e-1]
τms = [1e-7:1e-6:1e-5; 1e-5:1e-5:1e-3; 1e-3:1e-3:1.0; 1.0:0.2:100.0; 100.0:10.0:1000.0; 1e3:1e3:1e5; 1e5:1e5:1e7; ]
mzs = Array{Float64}(undef, length(τms), length(γ0s))

for i in eachindex(γ0s)
    for j in eachindex(τms)
        mzs[j, i]=sm.rho_en_relaxation(mx, my, mz, hx, hy, hz, hnorm*τms[j], γ0s[i]*τms[j])[3]
    end
end
```

```julia
Plots.plot(label=false)
labels = [L"\gamma_0=10^{-3}",L"\gamma_0=10^{-2}",L"\gamma_0=10^{-1}"]

for i in eachindex(γ0s)
    Plots.plot!(τms, mzs[:, i], linewidth=2.0, label=labels[i], xaxis=:log10, xlabelfontsize=30, ylabelfontsize=25,
            xtickfontsize=20, ytickfontsize=20, legend=:topright, legendfontsize=15, fontfamily="Computer Modern")
end

linestyles = [:dash, :dashdot, :dot]

for i in eachindex(γ0s)
    Plots.plot!([1/γ0s[i], 1/γ0s[i]], [minimum(mzs), maximum(mzs)], color="gray", linestyle=linestyles[i], linewidth=3.0, label=false)
    Plots.plot!([γ0s[i]/hnorm^2, γ0s[i]/hnorm^2], [minimum(mzs), maximum(mzs)], color="gray", linestyle=linestyles[i], linewidth=3.0, label=false) 
end

Plots.ylabel!(L"\langle M_{\mathrm{z}}\rangle")
Plots.xlabel!(L"\tau")
Plots.plot!(label=false)
```

```julia
hx=0.5
hy=0.1
hz=1.0
hnorm=sqrt(hx^2+hy^2+hz^2)

θ=acos(1/sqrt(3))
ϕ=π/4

mx=sin(θ)*cos(ϕ)
my=sin(θ)*sin(ϕ) 
mz=cos(θ) 

γ0 = 1e-2

ξs=[π/16, π/8, π/4, π/2]
τms = [1e-7:1e-6:1e-5; 1e-5:1e-5:1e-3; 1e-3:1e-3:1.0; 1.0:0.2:100.0; 100.0:10.0:1000.0; 1e3:1e3:1e5; 1e5:1e5:1e7; ]
#τms = [0.01:0.2:100.0; 100.0:10.0:1000.0; 1e3:1e3:1e5; 1e5:1e5:1e7; ]
entropy = Array{Float64}(undef, length(τms), length(ξs))


for i in eachindex(ξs)
    for j in eachindex(τms)
        entropy[j, i]=sm.entropy_weak(mx, my, mz, hx, hy, hz, γ0, τms[j], ξs[i])
    end
end
```

```julia
Plots.plot(label=false, size=(600, 400))
labels = [L"\xi=\pi/16", L"\xi=\pi/8", L"\xi=\pi/4", L"\xi=\pi/2"]

for i in eachindex(ξs)
    Plots.plot!(τms, entropy[:, i], linewidth=2.0, label=labels[i], xaxis=:log10, xlabelfontsize=25, ylabelfontsize=20,
            xtickfontsize=20, ytickfontsize=20, legend=:topright, legendfontsize=15, fontfamily="Computer Modern")
end


Plots.plot!([γ0/hnorm, γ0/hnorm], [minimum(entropy), maximum(entropy)], color="gray", linestyle=:dash, linewidth=3.0, label=false)
Plots.plot!([1/γ0, 1/γ0], [minimum(entropy), maximum(entropy)], color="gray", linestyle=:dash, linewidth=3.0, label=false)


Plots.ylabel!("Entropy")
Plots.xlabel!(L"\tau")
Plots.plot!(label=false)
```

```julia
θ = acos(1/sqrt(3))
ϕ = π/4
mx = sin(θ)*cos(ϕ)
my = sin(θ)*sin(ϕ)
mz = cos(θ)
γ0 = 0.01
hx = 0.1


ξs = [π/50, π/3, π/2]
Δs = [-3.01:0.01:3.01; ]
τms = [1e-7:1e-6:1e-5; 1e-5:1e-5:1e-3; 1e-3:1e-3:1.0; 1.0:0.2:100.0; 100.0:10.0:1000.0; 1e3:1e3:1e5; 1e5:1e5:1e7; ]
#τms = [100.0:10.0:10000.0; ]
rhoxs = Array{Float64}(undef, length(Δs), length(τms), length(ξs))
rhoys = Array{Float64}(undef, length(Δs), length(τms), length(ξs))
rhots = Array{Float64}(undef, length(Δs), length(τms), length(ξs))
rhozs = Array{Float64}(undef, length(Δs), length(τms), length(ξs))


for k in eachindex(ξs)
    for i in eachindex(τms)
        for j in eachindex(Δs)
            rhoxs[j, i, k] = sm.ss_matrix_weak(mx, my, mz, hx, 0.0, Δs[j]/2, γ0, τms[i], ξs[k])[1]
            rhoys[j, i, k] = sm.ss_matrix_weak(mx, my, mz, hx, 0.0, Δs[j]/2, γ0, τms[i], ξs[k])[2]
            rhozs[j, i, k] = sm.ss_matrix_weak(mx, my, mz, hx, 0.0, Δs[j]/2, γ0, τms[i], ξs[k])[3]
            rhots[j, i, k] = sqrt(rhoxs[j, i, k]^2+rhoys[j, i, k]^2)
        end
    end
end
```

```julia
f = Figure(size=(2200, 2000), fonts = (; regular = "cm", weird = "Blackchancery"))

columnlabels = [L"ξ=π/50", L"ξ=π/3", L"ξ=π/2"]
rowlabels = [L"\langle M_{\mathrm{x}}\rangle", L"\langle M_{\mathrm{y}}\rangle", L"\langle |M_{\perp}|\rangle", L"\langle M_{\mathrm{z}}\rangle"]

for k in eachindex(ξs)
    ax = Axis(f[1, 2*k-1], yscale=log10, xticklabelsize=40, yticklabelsize=40, titlesize=60, ylabelsize=60, title=columnlabels[k], ylabel="$(k==1 ? rowlabels[1] : "")")
    surf = CairoMakie.contourf!(Δs, τms, rhoxs[:, :, k], colormap=:coolwarm,
           levels=range(-maximum([-minimum(rhoxs[:, :, k]), maximum(rhoxs[:, :, k])]), maximum([-minimum(rhoxs[:, :, k]), maximum(rhoxs[:, :, k])]), length=15)) 
    Colorbar(f[1, 2*k], surf, ticklabelsize=40)
end

for k in eachindex(ξs)
    ax = Axis(f[2, 2*k-1], yscale=log10, xticklabelsize=40, yticklabelsize=40, ylabelsize=60, ylabel="$(k==1 ? rowlabels[2] : "")")
    surf = CairoMakie.contourf!(Δs, τms, rhoys[:, :, k], colormap=:coolwarm,
        levels=range(-maximum([-minimum(rhoys[:, :, k]), maximum(rhoys[:, :, k])]), maximum([-minimum(rhoys[:, :, k]), maximum(rhoys[:, :, k])]), length=15)) 
    Colorbar(f[2, 2*k], surf, ticklabelsize=40)
end

for k in eachindex(ξs)
    ax = Axis(f[3, 2*k-1], yscale=log10, xticklabelsize=40, yticklabelsize=40, ylabelsize=60, ylabel="$(k==1 ? rowlabels[3] : "")")
    surf = CairoMakie.contourf!(Δs, τms, rhots[:, :, k], colormap="Greys",
        levels=range(0.0, maximum([-minimum(rhots[:, :, k]), maximum(rhots[:, :, k])]), length=15))
    Colorbar(f[3, 2*k], surf, ticklabelsize=40)
end

for k in eachindex(ξs)
    ax = Axis(f[4, 2*k-1], yscale=log10, xticklabelsize=40, yticklabelsize=40, ylabelsize=60, xlabelsize=60,
        ylabel="$(k==1 ? rowlabels[4] : " ")", xlabel="$(k==2 ? "Δ" : " ")")
    surf = CairoMakie.contourf!(Δs, τms, rhozs[:, :, k], colormap=:coolwarm,
        levels=range(-maximum([-minimum(rhozs[:, :, k]), maximum(rhozs[:, :, k])]), maximum([-minimum(rhozs[:, :, k]), maximum(rhozs[:, :, k])]), length=15)) 
    Colorbar(f[4, 2*k], surf, ticklabelsize=40)
end

f
```
